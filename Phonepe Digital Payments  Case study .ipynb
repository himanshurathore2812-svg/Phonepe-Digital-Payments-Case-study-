```python
# Task 2: Exploratory Data Analysis (EDA) 
# 2.1: Analyze transaction trends over the years for each state

#1. Calculate the total number of transactions and total transaction amount for each state over the years. Display the results in a tabular format.

state_summary = state_txn_users.groupby(
    ['State','Year']
).agg({
    'Transactions':'sum',
    'Amount (INR)':'sum'
}).reset_index()

state_summary
```

```python
# 2. Identify the top 5 states with the highest transaction volumes and the top 5 states with the lowest transaction volumes. Display the results.

top5 = state_txn_users.groupby('State')['Transactions']\
.sum()\
.sort_values(ascending=False)\
.head()

top5  # highest
```

```python
# 2.2: Identify the most common transaction types in each state and quarter 
# 1. For each state and quarter, determine the most frequent transaction type. Display the results in a tabular format. 

idx = state_txnsplit.groupby(
    ['State','Quarter']
)['Transactions'].idxmax()

common_type = state_txnsplit.loc[
    idx,
    ['State','Quarter','Transaction Type','Transactions']
]

common_type
```

```python
# 2.4: Create a list of the top district per state based on population 
# 1. For each state, identify the district with the highest population. Display the results in a tabular format.

idx = district_demo.groupby(
    'State'
)['Population'].idxmax()

top_district = district_demo.loc[
    idx,
    ['State','District','Population']
]

top_district
```

```python
# Task 3: Data Quality Checks 
# 3.1: Ensure data consistency across state and district levels

# 1. For each state, calculate the total number of transactions, total transaction 
#amount, and total registered users by summing up the values from the 
#district level data.

district_summary = district_txn_users.groupby(
    ['State','Year','Quarter']
).agg({
    'Transactions':'sum',
    'Amount (INR)':'sum',
    'Registered Users':'sum'
}).reset_index()
```

```python
state_summary = state_txn_users.groupby(
    ['State','Year','Quarter']
).agg({
    'Transactions':'sum',
    'Amount (INR)':'sum',
    'Registered Users':'sum'
}).reset_index()
```

```python
# 2. Compare the results with the corresponding values at the state level to ensure they match. 

compare = district_summary.merge(
    state_summary,
    on=['State','Year','Quarter'],
    suffixes=('_district','_state')
)
```

```python
# 3. Display any discrepancies found between the district-level and state-level data. 

discrepancy = compare[
(compare['Transactions_district']!=compare['Transactions_state']) |
(compare['Amount (INR)_district']!=compare['Amount (INR)_state']) |
(compare['Registered Users_district']!=compare['Registered Users_state'])
]

discrepancy
```

```python
# Task 4: Data Merging and Advanced Analysis 
#4.1: Ratio of users to population by state 

#1. Merge the State_Txn and Users dataset with the District Demographics dataset to calculate the ratio of registered users to the population for each 
#state. 
# Display the results in a tabular format. 

state_population = district_demo.groupby(
    'State'
)['Population'].sum().reset_index()
```

```python
users = state_txn_users.groupby(
    'State'
)['Registered Users'].max().reset_index()

user_population = users.merge(
    state_population,
    on='State'
)
```

```python
user_population['User_to_Population_Ratio'] = (
user_population['Registered Users']
/
user_population['Population']
)
```

```python
# 2. Create a column chart depicting the ratio of users to population by state.

plt.figure(figsize=(15,6))

plt.bar(
    user_population['State'],
    user_population['User_to_Population_Ratio']
)

plt.xticks(rotation=90)

plt.show()
```

```python
# 4.2: Correlate population density with transaction volume 
# 1. Merge the District_Txn and Users dataset with the District Demographics dataset. 

merged = district_txn_users.merge(
    district_demo,
    on=['State','District','Code']
)
```

```python
# 2. Calculate the correlation between population density and transaction volume. 

merged[['Density','Transactions']].corr()
```

```python
# 3. Create a scatter plot to visualize the correlation between population density and transaction volume. 

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=merged,
    x='Density',
    y='Transactions'
)

plt.show()
```

```python
# 4.3: Average transaction amount per user 

#1. Merge relevant datasets to calculate the average transaction amount per user for each state. Display the results in a tabular format.

state_amt_user = state_txn_users.groupby(
    'State'
).agg({
    'Amount (INR)':'sum',
    'Registered Users':'max'
}).reset_index()
```

```python
state_amt_user['Amount_per_User'] = (
state_amt_user['Amount (INR)']
/
state_amt_user['Registered Users']
)

state_amt_user
```

```python
# 2. Identify the top 5 states with the highest average transaction amount per user and the top 5 states with the lowest average transaction amount per 
# user. Display the results. 

state_amt_user.sort_values(
    by='Amount_per_User',
    ascending=False
).head()   # highest
```

```python
state_amt_user.sort_values(
    by='Amount_per_User'
).head()   # Lowest
```

```python
# 4.4: Device brand usage ratio 
# 1. Merge the State_DeviceData dataset with the State_Txn and Users dataset. 

brand_ratio = state_device.merge(
    users,
    on='State',
    suffixes=('_brand','_total')
)
```

```python
# 2. Calculate the ratio of users using each device brand to the total number of registered users in each state. Display the results in a tabular format.

brand_ratio['Brand_Ratio'] = (
brand_ratio['Registered Users_brand']
/
brand_ratio['Registered Users_total']
)
```

```python
# 3. Create a bar chart depicting the device brand usage ratio for each state. 


maharashtra = brand_ratio[
    brand_ratio['State'] == 'Maharashtra'
]

plt.figure(figsize=(10,5))

sns.barplot(
    data=maharashtra,
    x='Brand',
    y='Brand_Ratio'
)

plt.xticks(rotation=45)

plt.title('Device Brand Usage Ratio in Maharashtra')
plt.xlabel('Brand')
plt.ylabel('Usage Ratio')

plt.show()
```
